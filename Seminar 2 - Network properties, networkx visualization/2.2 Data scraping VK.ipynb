```python
def get_my_friends_ids(user_id, token=-1): 
    if token == -1:    
        token   = _token
        
    friends_url = f'https://api.vk.com/method/friends.get?user_ids={user_id}&fields=bdate&access_token={token}&v=5.21' 
    # также вы можете добавить access_token в запрос, получив его через OAuth 2.0
    json_response = requests.get(friends_url.format(user_id)).json()
    if json_response.get('error'):
        #print json_response.get('error')
        return list()
    return json_response['response']['items']

def get_mutel_friends_ids(user_id, friend_id, token=-1): 
    if token == -1:    
        token   = _token
        
    friends_url = f'https://api.vk.com/method/friends.getMutual?source_uid={user_id}&target_uid={friend_id}&fields=bdate&access_token={token}&v=5.21' 
    # также вы можете добавить access_token в запрос, получив его через OAuth 2.0
    json_response = requests.get(friends_url.format(user_id)).json()
    if json_response.get('error'):
        #print json_response.get('error')
        return list()
    return json_response['response']

def get_user_information(user_id, token=-1):
    if token == -1:    
        token = _token
    
    friends_name  = f"https://api.vk.com/method/users.get?user_ids={user_id}&fields=bdate&access_token={token}&v=5.21"
    json_response = requests.get(friends_name.format(user_id)).json()
    if json_response.get('error'):
        #print json_response.get('error')
        return list()
    return json_response['response'][0]

def get_friend_information(user_id, token=-1):
    if token == -1:    
        token = _token
    
    friends_name  = f"https://api.vk.com/method/friends.get?user_ids={user_id}&fields=city&access_token={token}&v=5.21"
    json_response = requests.get(friends_name.format(user_id)).json()
    if json_response.get('error'):
        #print json_response.get('error')
        return list()
    return json_response['response']

def get_friend_information_education(user_id, token=-1):
    if token == -1:    
        token = _token
    
    friends_name  = f"https://api.vk.com/method/friends.get?user_ids={user_id}&fields=education&access_token={token}&v=5.21"
    json_response = requests.get(friends_name.format(user_id)).json()
    if json_response.get('error'):
        #print json_response.get('error')
        return list()
    return json_response['response']
```
