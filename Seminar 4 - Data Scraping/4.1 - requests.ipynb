# Simplest scraping
**We gonna show how to open a very simple [page](https://anr.hse.ru/), and read all the content in it.**

```python
import requests

html = requests.get("https://www.hse.ru/news/admission/")

#with open("hse.html", "w") as fp:
#    fp.write(html.text)
    
print(html.text)
```

```python
import re

res = re.findall(r"<title>(.+?)</title>", html.text)
print("\nPage title is: ", res[0])
```

**Then we select some text according to the tag arround text using regular expression**

```python
import re

res = re.findall(r"(?<=class=\"link link_dark2 no-visited\">)(.*?)(?=</a>)", html.text)
for hr in res:
    print(f"### {hr}")
```

**Another example to select paragraph content out of html.**

**And select links using regex**

```python
res = re.findall(r'href="(.*?\.)', html.text)
for hr in res:
    print(f"### {hr}")
```
