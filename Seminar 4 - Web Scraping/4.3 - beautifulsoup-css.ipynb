# Beautifulsoup: find by CSS class 
**First we still need to open a [page](https://anr.hse.ru/), then we can apply beautifulsoup on this page's html.**

```python
from bs4 import BeautifulSoup
import requests


resp = requests.get("https://www.hse.ru/news/admission/")
html = resp.text
```

**Parsing this html using a method called lxml, create a soup object. Find all "li" tag which has a class=month.**

```python
soup = BeautifulSoup(html, 'lxml')

# use class to narrow search
posts = soup.find_all('div', {"class": "post"})
for post in posts:
    print("####\n" + post.get_text())
```

**Or using some helpful functions to find tags. Access the attribute of found tags using a key just like what you do in a python dictionary.**

```python

posts = soup.find_all('div', {"class": "post"})
for post in posts:
    try:
        post_content = post.find('div', {"class": "post__text"})
        print("####\n" + post_content.get_text())
    except:
        pass
```
