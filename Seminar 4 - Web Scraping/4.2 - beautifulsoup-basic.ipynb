# Basic usage of beautifulsoup
**First we still need to open a [page](https://anr.hse.ru/), then we can apply beautifulsoup on this page's html.**

```python
from bs4 import BeautifulSoup
import requests

resp = requests.get("https://www.hse.ru/news/admission/")
html = resp.text
#print(html)
```

**Parsing this html using a method called html.parser, create a soup object. You can simply "h1" or "p" to call the heading 1 and paragraph tag from soup.**

```python
soup = BeautifulSoup(html, "html.parser")
print(soup.a)
#print('\n', soup.p)
```

**Or using some helpful functions to find tags. Access the attribute of found tags using a key just like what you do in a python dictionary.**

```python
all_href = soup.find_all('a')
all_href = [l['href'] for l in all_href]
for hr in all_href:
    print(f"### {hr}")
```
